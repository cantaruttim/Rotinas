# Regras de Associação
- Não podem ser dados numéricos

```python
dataset_apriori = df[['workclass', 'education', 'marital-status', 'relationship', 'occupation',
                            'sex', 'native-country', 'income']]
```

```python
dataset_apriori.head()
```

```python
dataset_apriori = dataset_apriori.sample(n = 1000, random_state=42)
dataset_apriori.shape
```

```python
## Amostra aleatória simples selecionada

dataset_apriori.head()
```

```python
transacoes = []
for i in range(dataset_apriori.shape[0]):
  transacoes.append([str(dataset_apriori.values[i, j]) for j in range(dataset_apriori.shape[1])])
```

```python
transacoes[:2]
```

```python
!pip install apyori
```

```python
from apyori import apriori

regras = apriori(transacoes, min_support = 0.3, min_confidence = 0.2)
resultados = list(regras)
```
