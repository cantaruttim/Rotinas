```python
import tabula
```

```python
## Podemos ver que na página que queremos trabalhar desse documento, temos duas tabelas

lista_tabelas = tabula.read_pdf("https://hospitalangelinacaron.org.br/hac/wp-content/uploads/2022/12/Balanco-Social-2021.pdf", pages='8')
print(len(lista_tabelas))
```

```python
## Podemos visualizar as tabelas
### Obs¹: Ao realizar a leitura, podemos ver que as tabelas estão unificadas (nos mostrando apenas 1 tabela)

for tabela in lista_tabelas:
  display(tabela)
```

```python
## SEPARANDO AS DUAS TABELAS DA TABELA SELECIONADA

tabelas = lista_tabelas[0]

## Selecionando as linhas que nós queremos apenas (no caso a primeira tabela estava entre as linhas de indice 0 e 25)
tabela = tabelas[0:25]

## Para separar as colunas da outra tabela, realizei um drop para separá-los
tabela = tabela.drop(columns = ["Unnamed: 1", 'SUS- Incentivo - Diárias Extras de UTI', 'Unnamed: 2', '9.959', '136.865'], axis = 1)

## Nessa tabela tinhamos duas colunas de ano com os valores, então realizei o split para poder separá-las
tabela[['2021', '2022']] = tabela["Unnamed: 0"].str.split(" ", expand=True)

## Retirei a coluna vazia que não tinha puxado os valores
tabela = tabela.drop("Unnamed: 0", axis = 1)

display(tabela)
```

```python
## Primeira Tabela salva em xlsx
tabela.to_excel("Tabelas.xlsx")
```

```python
# Realizei uma nova leitura do arquivo (para pegarmos ele no formato original)

lista_tabelas = tabula.read_pdf("https://hospitalangelinacaron.org.br/hac/wp-content/uploads/2022/12/Balanco-Social-2021.pdf", pages='8')
print(len(lista_tabelas))
```

```python
# Creio que não havia mencionado mas as tabelas com o tabula são convertidas em listas de tabelas, as quais precisam
# dessa loop `for` para a visualização

for tabela in lista_tabelas:
  display(tabela)
```

```python
## SEPARANDO AS DUAS TABELAS DA TABELA SELECIONADA

tabelas = lista_tabelas[0]

# A primeira parte da segunda tabela encontrava-se entre os indices 32 e 46
tabela = tabelas[32:46]

# Retiramos as colunas que faziam parte da primeira tabela.
tabela = tabela.drop(columns = ["Unnamed: 1", 'SUS- Incentivo - Diárias Extras de UTI', 'Unnamed: 2', '9.959', '136.865'], axis = 1)

# Realizamos o mesmo processo de split dos valores
tabela[['2021', '2022']] = tabela["Unnamed: 0"].str.split(" ", expand=True)

# Retiramos a coluna que não nos interessa
tabela = tabela.drop("Unnamed: 0", axis = 1)


display(tabela)
```

```python
tabela.to_excel("Continuação.xlsx")
```
